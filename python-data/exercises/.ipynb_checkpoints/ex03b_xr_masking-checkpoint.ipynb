Q1. For masking, we're back to using our ocean dataset. Load it now from `../data/xbhubo.pgc0apr.nc`.

```python
import xarray as xr

ds = xr.open_dataset('../data/xbhubo.pgc0apr.nc')

temperature = ds["temp"]
```

Q2. Determine which grid cells sea surface temperaturevis more than the mean.

```python
# Get the surface dataset for all variables
surf = ds.sel(depth=0, method='nearest')
# Specify the mask for temperature only (that's why you have to do surf.temp; you're specifing the variable you want
surf.temp.where(surf.temp > surf.temp.mean()).plot()
```

Q3. Combine this with another .where() lookup to show only cells where the temperature is more than the mean and salinity is more than the mean.

```python
# Now we want 2 conditions for the mask, which we can just do one after the other
surf.temp.where(surf.temp > surf.temp.mean()).where(surf.salinity > surf.salinity.mean()).plot()
```

Q4. Masks are just boolean arrays. Create a re-usuable mask for the temperature and sst criteia above, and a combined one.

```python
temp_mask = surf.temp > surf.temp.mean()
salin_mask = surf.salinity > surf.salinity.mean()
temp_salin_mask = temp_mask & salin_mask
```

Q5. Use this mask to make the same temperature plot, and a similar one for sst.

```python
# Use the combined mask
surf.temp.where(temp_salin_mask).plot()
```

```python
# Do a plot for just the salinity
surf.salinity.where(salin_mask).plot()
```
